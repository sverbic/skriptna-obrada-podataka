## Koroplet

Prvi primer vizuelizacije u ovoj svesci je tipa koroplet (eng. _choropleth_). Kod ovog prikaza teritoriju na mapi delimo na manje delove sa jasno određenim konturama pa oblasti unutar kontura bojimo u skladu sa veličinom koju želimo da prikažemo za taj deo teritorije. Mi ćemo koristeći već poznate podatke Republičkog zavoda za statistiku koropletom prikazati prosečnu neto zaradu po opštinama. 

```python
import geopandas as gpd
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Fajl koji nam je neophodan za prikazivanje mapa je zbirka kontura oblasti koje hoćemo da prikažemo. Te konture su date kao nizovi koordinata, odnosno poligoni. U tabeli koju smo preuzeli sa sajta Republičkog geodetskog zavoda za svaku opštinu imamo podatke o poligonima u dva formata: _wkt_ i _geometry_. Ta se tabela nalazi u fajlu __opstina.csv__. Pošto su imena opština data i ćirilicom obavezno stavite `encoding='UTF-8'`. Obratite pažnju da csv fajl otvaramo pomoću funkcije `read_file()` biblioteke __geopandas__.

```python
konture=gpd.read_file('data/opstina.csv', encoding='UTF-8')
```

Struktura __konture__ jeste tabela, ali pripada posebnom tipu podataka sa kojim se do sada nismo susretali. Ukoliko učitana tabela sadrži kolonu __geometry__ sa poligonima, Pajton će pomoću funkcije `plot()` znati da ih iscrta čak i bez ikakvih argumenata.

```python
type(konture)
```

```python
konture.head(2)
```

Ovako izgledaju konture svih 197 opština u Srbiji.

```python
konture.plot()
```

Primetite da koordinate na mapi ne odgovaraju stepenima geografske dužine i širine. Razlog za to je što se u geodeziji više koriste [UTM koordinate](https://en.wikipedia.org/wiki/Universal_Transverse_Mercator_coordinate_system). Ko hoće može da izvrši konverziju ovih podataka, ali nama sada to nije potrebno. Hoćemo samo mapu. Štaviše, mogli bismo da isključimo okvir jer nam ovde praktično ničemu ne služi.

### Povezivanje kontura sa vrednostima

Podatke o neto zaradi i konturama opština smo našli u dva različita fajla. Sada je potrebno da povežemo te dve tabele. Za početak, da učitamo i tu drugu tabelu.

```python
opps=pd.read_csv("data/03IND01.csv",sep=";")
```

```python
opps.head()
```

Ova tabela ima više podataka nego što nam je potrebno. Takođe, u istoj koloni ima različite indikatore. Izdvojićemo samo indikator koji nas interesuje (neto zarada) i ukloniti nepotrebne kolone.

```python
neto_zarada=opps[opps['Indikator']=='Просечна нето зарада за месец, према општини пребивалишта запослених']
```

```python
neto_zarada.head()
```

```python
neto_zarada=neto_zarada[['idter','nter','vrednost']]
```

```python
neto_zarada = neto_zarada.astype({'vrednost':float})
```

```python
neto_zarada
```

Ako pažljivo pogledate nazivi opština nisu isti u prvoj i drugoj tabeli. U prvoj (__konture__), na primer, piše "ALEKSANDROVAC" i "АЛЕКСАНДРОВАЦ", dok u drugoj (__neto_zarada__) imamo "Александровац". Povezivanje po imenu opštine bi moglo da bude problematično. Srećom, u obe tabele imamo matične brojeve opština. U prvoj tabeli to je kolona __opstina_maticni_broj__, a u drugoj __idter__. Da bi dalji rad bio jednostavniji, pomoću funkcije `rename()` preimenovaćemo naziv kolone u obe tabele u __MBO__.

```python
konture=konture.rename(columns={'opstina_maticni_broj':'MBO'})
```

```python
konture['MBO']=konture['MBO'].astype(int)
```

```python
neto_zarada=neto_zarada.rename(columns={'idter':'MBO'})
```

```python
neto_zarada['MBO']=neto_zarada['MBO'].astype(int)
```

Povezivanje dve tabele ćemo uraditi pomoću funkcije `merge()` po koloni __MBO__.

```python
konture2 = pd.merge(konture,neto_zarada,on='MBO')
```

```python
konture2.head()
```

Sada bi za svaku opštinu u tabeli trebalo da imamo i podatke o zaradi i konture. Ostalo je još samo da to prikažemo na mapi tako što ćemo kao argument funkcije `plot()` staviti neto zaradu, odnosno `column='vrednost'`.

```python
konture2.plot(column='vrednost',legend=True)
```

Kada pažljivo pogledate mapu, videćete da podaci za opštine na Kosovu i Metohiji nisu prikazani. Tabela opština sa konturama ima 197 redova, a tabela sa podacima o zaradama 175. Republički zavod za statistiku u svojim izveštajima nema mesečne pokazatelje za KiM. Zbog toga je spojena tabela kraća za ta 24 reda. Imajte u vidu da zvanični podaci različitih državnih organa mogu da se razlikuju po ovom pitanju.

Što se estetike mape tiče, istu mapu možemo da prikažemo i sa nešto više detalja. U narednoj ćeliji je primer kako to može da se uradi. Slobodno varirajte parametre i izaberite mapu koja se vama najviše sviđa.

```python
konture2.plot(column='vrednost', legend=True, cmap="hot", figsize=(15, 10))
plt.axis('off')
plt.title('Prosečna mesečna neto zarada po opštinama', fontdict={'fontsize': '20'})

# navođenje izvora podatka
plt.annotate('Izvori podataka: \nhttp://data.gov.rs (2020)\nhttps://opendata.geosrbija.rs (2021)',
             xy=(0.2, 0.10),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', 
             fontsize=10, color='#555555');
```

## Prikaz geolokacija

Mnogi podaci koje imamo u tabelama predstavljaju geo-podatke, odnosno podatke u vezi sa raznim toponimima, npr. državama, opštinama ili lokacijama. To nam omogućava da određene podatke pridružimo geo-lokacijama i predstavimo ih na mapi.

Lokacije objekata date sa geografskom dužinom (longitudom) i širinom (latitudom) nazivamo geolokacijama. Fajlovi u kojima se navode geolokacije mogu imati različite formate (CSV, JSON, KML itd.). Koji god da uzmemo, Pajton će znati kako da ga prevede u _DataFrame_. 

Ovde ćemo konkretno uzeti [lokacije državnih automatskih mernih stanica za praćenje kvaliteta vazduha u Srbiji](https://data.gov.rs/sr/datasets/metapodatsi-drzhavne-mrezhe-automatskikh-mernikh-stanitsa-za-pratshenje-kvaliteta-vazdukha/). Videćemo tačno koliko ima ovakvih stanica i gde se nalaze. CSV fajl sa ovim podacima je dostupan na Portalu otvorenih podataka. Kopija fajla je u folderu __data__ sa ostalim relevantnim podacima za ovaj praktikum.  

```python
data=pd.read_csv("data/stanice_kvalitet_vazuha.csv")
```

```python
data.head()
```

Za prikaz ovih podataka iskoristićemo biblioteku __folium__ koja zahteva pristup internetu kako bi funkcije preuzele odgovarajuće mape. Ukoliko nemate instaliranu ovu biblioteku, samo učinite to sa `pip install folium`. 

```python
import folium
```

Mapa koju kreira funkcija `Map()` je objekat čiji su argumenti centar mape, tip mape i razmera. Dovoljno je da napišemo ime ovog objekta u ćeliji koju izvršavamo da dobijemo detaljnu interaktivnu mapu koju proizvoljno možemo da pomeramo i zimiramo.

```python
mapa = folium.Map(location=[44,20], tiles="OpenStreetMap", zoom_start=7)
```

```python
mapa
```

Dodavanje lokacija na koje ćemo staviti markere radimo pomoću funkcije `Marker()` čiji su argumenti latituda, longituda i oznaka lokacije. Te detalje onda dodajemo mapi pomoću funkcije `add_to()`.  

```python
for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['latitude'], data.iloc[i]['longitude']],
      popup=data.iloc[i]['naziv'],
   ).add_to(mapa)
```

```python
mapa
```

Interaktivnu mapu konačno možemo da snimimo u html fajl pomoću funkcije `save()`.

```python
mapa.save('output/automatske_merne_stanice.html')
```
